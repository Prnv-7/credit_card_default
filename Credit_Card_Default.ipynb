```python
x = df.drop(columns = ['Defaulted','ID', 'LIMIT_BAL', 'SEX', 'MARRIAGE', 'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6', 'PAY_AMT1','PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5','PAY_AMT6'])
```

### Here, we have dropped every other columns which are not important for building the model. To identify the unwanted columns we plotted a heatmap of the correlation of the dataframe. From that plot we can pretty much understand which all features will least affect the prediction - Feature Selection

```python
y = df['Defaulted']
```

### Scaling of data is done by using StandardScaler. Here only x is scaled.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
X = pd.DataFrame(x_scaled, columns = x.columns)
```

```python
X.head()
```

### Now let's split the data into training data and testing data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

## Using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```
